#***Models***
#***1.Logistic Regression***

```python
df
```

```python
X=df.iloc[:,:13]
X
```

```python
Y=df.iloc[:,13:]
Y
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,plot_confusion_matrix,accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
Y_train=np.ravel(Y_train)
clf=LogisticRegression()
clf.fit(X_train,Y_train)
pred=clf.predict(X_test)
log_acc=accuracy_score(pred,Y_test);
print("REPORT: ")
print(confusion_matrix(Y_test,pred))
print(classification_report(Y_test,pred))
plot_confusion_matrix(clf,X_test,Y_test)
plt.grid(False)
plt.gcf().set_size_inches(12,6)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf=KNeighborsClassifier(n_neighbors=9)
clf.fit(X_train,Y_train)
pred=clf.predict(X_test)
knn=accuracy_score(pred,Y_test);
print("REPORT: ")
print(confusion_matrix(Y_test,pred))
print(classification_report(Y_test,pred))
plot_confusion_matrix(clf,X_test,Y_test)
plt.grid(False)
plt.gcf().set_size_inches(12,6)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier(criterion="entropy", max_depth=3)
clf.fit(X_train,Y_train)
pred=clf.predict(X_test)
dtc=accuracy_score(pred,Y_test);
print("REPORT: ")
print(confusion_matrix(Y_test,pred))
print(classification_report(Y_test,pred))
plot_confusion_matrix(clf,X_test,Y_test)
plt.grid(False)
plt.gcf().set_size_inches(12,6)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators = 100)
clf.fit(X_train,Y_train)
pred=clf.predict(X_test)
rcf=accuracy_score(pred,Y_test);
print("REPORT: ")
print(confusion_matrix(Y_test,pred))
print(classification_report(Y_test,pred))
plot_confusion_matrix(clf,X_test,Y_test)
plt.grid(False)
plt.gcf().set_size_inches(12,6)
  
```

#***Report of Models***

```python
print("Accuracy of Logistic Regression : ",log_acc)
print("Accuracy of Decision Tree Classifier : ",dtc)
print("Accuracy of KNN Classifier : ",knn)
print("Accuracy of Random Forest Classifier : ",rcf)
models_acc=[log_acc,dtc,knn,rcf]
names_of_models=['LogisticRegression','DecisionTreeClassifier','KNNClassifier','RandomForestClassifier'];
sns.barplot(y=names_of_models,x=models_acc)
plt.gcf().set_size_inches(8,4)
plt.xlim([0.6,1.0])
```
