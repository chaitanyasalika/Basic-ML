#***IMPORTING LIBARIES***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

#***READ THE DATASET ("CORRUPTION 2019")***

```python
corruption = pd.read_csv("Corruption 2019.csv")
```

#***COLUMS OF THE DATASET("CORRUPTION2019")***

```python
corruption.columns
```

#***GETTING INFORMATION OF DATASET("CORRUPTION2019")***

```python
corruption.info()
```

```python
corruption.drop(["Category", "Unnamed: 11", "Unnamed: 12", "Unnamed: 13", "Unnamed: 14"], axis=1, inplace=True)
```

```python
corruption.head(30)
```

```python
corruption.isnull()
```

```python
plt.figure(figsize=(20,12))
plt.xticks(rotation=90)
sns.heatmap(corruption.isnull())
```

#***COUNTPLOT***

```python
plt.figure(figsize=(20,12))
plt.xticks(rotation=90)
sns.countplot(x='Total cases',data=corruption,palette='BuGn')
```

#***PAIRPLOT***

```python
sns.pairplot(corruption)
```

```python
totalcase = corruption.sort_values("Total cases", ascending = False)
```

#***POINT PLOT***

```python
plt.figure(figsize=(20,14))
sns.pointplot(data=totalcase, x='State/UT ', y='Total cases', color='palegreen');
plt.title("Total Case of Corruption by state")
plt.tick_params(axis='x', which='major', labelsize=15, rotation=90)
plt.tick_params(axis='y', which='major', labelsize=15)
```

```python
chargesheet = corruption.sort_values("Cases Charge-sheeted", ascending = False)
```

#***POINT PLOT***

```python
plt.figure(figsize=(20,12))
plt.xticks(rotation=90)
sns.pointplot(x='State/UT ', y='Cases Charge-sheeted', data=chargesheet, color='red');
```

```python
pending = corruption.sort_values("Cases Pending Investigation at End of the Year", ascending = False)
```

#***BAR PLOT***

```python
plt.figure(figsize=(20,12))
plt.xticks(rotation=90)
sns.barplot(x='State/UT ', y='Cases Pending Investigation at End of the Year', data=pending, color='red');
```

```python
previous = corruption.sort_values("Cases Pending Investigation from Previous Year", ascending = False)
```

```python
plt.figure(figsize=(20,12))
plt.xticks(rotation=90)
sns.pointplot(x='State/UT ', y='Cases Pending Investigation from Previous Year', data=previous, color='red');
```

```python
reported = corruption.sort_values("Cases Reported during the year", ascending = False)
```

```python
plt.figure(figsize=(20,12))
plt.xticks(rotation=90)
sns.pointplot(x='State/UT ', y='Cases Reported during the year', data=reported, color='black');
```

#***LINE PLOT***

```python
plt.figure(figsize=(20,12))
plt.xticks(rotation=90)
sns.lineplot(x='State/UT ', y='Total cases', data=reported, color='green');
```

#***DIST PLOT***

```python
plt.figure(figsize=(20,12))
plt.xticks(rotation=90)
sns.distplot(corruption['Cases Pending Investigation from Previous Year'].dropna(),color='darkviolet',bins=8)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='Cases Reported during the year',hue='Total cases',data=corruption,palette='BuGn')
```

#***BOX PLOT***

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Cases Pending Investigation from Previous Year',data=corruption,palette='BuGn')
```

#***SCATTER PLOT***

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Cases in Which Final Reports Submitted',y='Total cases',data=corruption,palette='BuGn')
```
